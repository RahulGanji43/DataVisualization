```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.random import randn, randint, uniform, sample
%matplotlib inline
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
fmri.describe()
```

```python
sns.relplot(x='timepoint',y='signal',size='event',hue='event',data=fmri,kind='line')
plt.title('Relation between timpoint and signal')
plt.show()
```

```python
sns.relplot(y='signal',x='timepoint',kind='line',hue='subject',size='region',data=fmri)
```

```python
sns.relplot(data=fmri,x='timepoint',y='signal',hue='region')
plt.title('Relation b/w timepoint and signal')
plt.show()
```

```python
sns.boxplot(data=fmri,x='timepoint',y='signal',hue='region')
plt.title('Relation b/w timepoint and signal')
plt.show()
```

```python
sns.boxplot(data=fmri,x='timepoint',y='signal',hue='event')
plt.title('Relation b/w timepoint and signal')
plt.show()
```

```python
sns.lineplot(y='signal',x='timepoint',data= fmri)
```

```python
sns.lineplot(y='signal',x='timepoint',data= fmri,hue='event')
```

```python
sns.lineplot(y='signal',x='timepoint',data= fmri,hue='region',style='event')
```

```python
sns.lineplot(data=fmri,x="timepoint", y="signal", hue="event", style="event",markers=True, dashes=False)
```
