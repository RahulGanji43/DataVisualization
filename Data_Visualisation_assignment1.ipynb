```python
import numpy as np
import pandas as pd
import seaborn as sns
from numpy.random import randn, randint, uniform, sample

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 #Question1
a = np.arange(40,50)
b = np.arange(50,60)
plt.plot(a,b)
plt.title('Line plot',color = 'blue')
plt.xlabel('x-axis', color ='green')
plt.ylabel('y-axis', color = 'green')
```

```python
#Question2
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
plt.plot(days,sales_1,'go-',linestyle='dashed',linewidth=1,markersize=10,label='company 1')
plt.plot(days,sales_2,'bo-',linewidth=1,markersize=10,label='company 2')
plt.title('Sales of a company')
plt.xlabel('Days of week')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
#Question 3
x=[1,2,3,4]
y1=[4,3,2,1]
y2=[10,20,30,49]
y3=[40,30,20,10]
y4=[1,2,1,2]
y5=[40,70,40,70]
```

```python
plt.subplot(3,3,1)
plt.plot(x,y1,'go',linestyle='dashed',linewidth=1,markersize=10)
plt.title('subplot 1',color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.subplot(3,3,2)
plt.plot(x,y2,'yo',linestyle='dashed',linewidth=1,markersize=10)
plt.title('subplot 2',color='yellow')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.subplot(3,3,3)
plt.plot(x,y3,'ro',linestyle='dashed',linewidth=1,markersize=10)
plt.title('subplot 3',color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.subplot(3,3,4)
plt.plot(x,y4,'bo',linestyle='dashed',linewidth=1,markersize=10)
plt.title('subplot 4',color='blue')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.subplot(3,3,5)
plt.plot(x,y5,'ko',linestyle='dashed',linewidth=1,markersize=10)
plt.title('subplot 5',color='black')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```
